### Limpieza de datos

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/ricardoahumada/DataScienceBasics/refs/heads/main/data/housing.csv')
df.info()
df.describe()
```

```python
df.isna().sum()
```

```python
dfdel = df.dropna()
dfdel.info()
```

```python
media = df['total_bedrooms'].mean # Sustitución vacíos con media
print(media)
df['total_bedrooms'].fillna(media, inplace=True)
df.info()
df.describe
```

```python
media = df['total_bedrooms'].median #Sustitución vacíos con mediana
print(media)
df['total_bedrooms'].fillna(media, inplace=True)
df.info()
df.describe
```

### Valores atípicos

```python
#Datos de muestra y adición de outliers

import numpy as np

np.random.seed(0)
data = np.random.randint(low=0, high=11, size=1000)
data[0] = 100
data[1] = -100
```

```python
#Calcular los z-scores (https://en.wikipedia.org/wiki/Standard_score) Identificar valores atípicos según z-cores

Z_score = (data-np.mean(data))/np.std(data)
umbral = 3
outliers = np.where(np.abs(Z_score) > umbral)
data[outliers]
```

Otro método consiste en calcular el rango intercuartílico (IQR: https://en.wikipedia.org/wiki/Interquartile_range) de la
distribución y clasificar cualquier valor que sea Q1-(1,5 x IQR) o Q3 + (1,5 x IQR) como valores atípicos potenciales

```python
# Calcular IQR
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1
print('iqr:', iqr)
print(np.mean(data), q1, q3)

# Identificar outliers según IQR
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
print('umbrales:', lower_bound, upper_bound)
#outliers = np.where((data < lower_bound) | (data > upper_bound))[0]
#print(data[outliers])
```
